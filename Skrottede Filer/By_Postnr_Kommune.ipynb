```python
import requests 
import json
import tqdm
import pandas as pd

url = 'https://dawa.aws.dk/supplerendebynavne2' 
response = requests.get(url, 'html.parsing').json()
```

```python
COL = ['Kommune', 'By', 'Postnr1', 'Postnr2', 'Postnr3', 'Postnr4']

df = pd.DataFrame(columns = COL)

for i in tqdm.tqdm(response):
    try:
        df = df.append({'Kommune': i['kommune']['navn'],
                       'By': i['navn'],
                       'Postnr1': i['postnumre'][0]['nr'],
                       'Postnr2': i['postnumre'][1]['nr'],
                       'Postnr3': i['postnumre'][2]['nr'],
                       'Postnr4': i['postnumre'][3]['nr']}, ignore_index = True)
        
    except: 
        try:
            df = df.append({'Kommune': i['kommune']['navn'],
                            'By': i['navn'],
                            'Postnr1': i['postnumre'][0]['nr'],
                            'Postnr2': i['postnumre'][1]['nr'],
                            'Postnr3': i['postnumre'][2]['nr']}, ignore_index = True)
        except: 
            try:
                df = df.append({'Kommune': i['kommune']['navn'],
                                    'By': i['navn'],
                                    'Postnr1': i['postnumre'][0]['nr'],
                                    'Postnr2': i['postnumre'][1]['nr']}, ignore_index = True)
            except:
                try:
                    df = df.append({'Kommune': i['kommune']['navn'],
                                    'By': i['navn'],
                                    'Postnr1': i['postnumre'][0]['nr']}, ignore_index = True)
                except:
                    None
                    
```
