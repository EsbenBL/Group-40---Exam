```python
import requests 
import json
url = 'https://dawa.aws.dk/postnumre'
response = requests.get(url).json()
```

```python
response[0]
```

```python
import tqdm
import pandas as pd
COL = ['Kommune', 'Postnr']
df = pd.DataFrame(columns = COL)

for i in tqdm.tqdm(response):
    df = df.append({'Kommune': i['adgangsadresse']['kommune']['navn'],
                    'Postnr': i['adgangsadresse']['postnummer']['nr']}, ignore_index=True)
```
