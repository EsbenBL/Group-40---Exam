```python
import requests
import json

url = 'http://dawa.aws.dk//adresser?retskredskode=1115'

response = requests.get(url)
response_json = response.json()
```

```python
#TEST!

import tqdm
import pandas as pd
COL = ['Region', 'Kommune', 'Postnr', 'Zone']
df = pd.DataFrame(columns = COL)

for i in tqdm.tqdm(range(100)):
    split = response_json[i]
    df = df.append({'Region': split['adgangsadresse']['region']['navn'], 
                   'Kommune': split['adgangsadresse']['kommune']['navn'],
                    'Postnr': split['adgangsadresse']['postnummer']['nr'],
                    'Zone' : split['adgangsadresse']['zone']}, ignore_index=True)
```
