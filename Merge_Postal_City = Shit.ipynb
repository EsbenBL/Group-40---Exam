```python
import pandas as pd
pd.options.display.max_columns = 100

df_house = pd.read_csv('housing_data_with_street_address_split.csv')
df_kommune = pd.read_csv('By_Postnr_Kommune.csv')
df_kommune = df_kommune.drop(columns = [	'Unnamed: 0'])
```

```python
#Make a full adress column for the DAWA url connect to. 
df_house['addr_full'] = df_house['address'] + ' ' + df_house['city']
```

```python
import requests

def get_what_we_want(addr):
    url = lambda q: 'https://dawa.aws.dk/adresser?q={}'.format(q)
    resp = requests.get(url(addr))
    
    if not resp.status_code == 200:
        return ''
    resp = resp.json()
    
    pnr = resp[0]['adgangsadresse']['adgangspunkt']['koordinater']
    
    return pnr
    
```

```python
get_what_we_want(df_house['addr_full'][0])
```
