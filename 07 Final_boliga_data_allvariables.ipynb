```python
from ast import literal_eval

poitypes = ["lake", "forest", "doctor", "supermarket", "school", "daycare", "hospital", "metro",
            "train", "stop", "strain", "junction", "pharmacy", "library", "coast", "airport"]
def get_lakekey(key_column):
    key_column = literal_eval(key_column)
    return key_column['lake']
def get_forestkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['forest']
def get_doctorkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['doctor']
def get_supermarketkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['supermarket']
def get_schoolkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['school']
def get_daycarekey(key_column):
    key_column = literal_eval(key_column)
    return key_column['daycare']
def get_hospitalkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['hospital']
def get_metrokey(key_column):
    key_column = literal_eval(key_column)
    return key_column['metro']
def get_trainkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['train']
def get_stopkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['stop']
def get_strainkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['strain']
def get_junctionkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['junction']
def get_pharmacykey(key_column):
    key_column = literal_eval(key_column)
    return key_column['pharmacy']
def get_librarykey(key_column):
    key_column = literal_eval(key_column)
    return key_column['library']
def get_coastkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['coast']
def get_airportkey(key_column):
    key_column = literal_eval(key_column)
    return key_column['airport']
```

```python
finaldf['lake_distance'] = finaldf['hvorlangt_total'].apply(get_lakekey)
```
