```python
def xtract_proptype(code):
    if int(code) == 500:
        return 'Andelsbolig'
    elif int(code) == 300:
        return 'Ejerlejlighed'
    elif int(code) == 400:
        return 'Fritisbolig'
    elif int(code) == 800:
        return 'Fritidsgrund'
    elif int(code) == 700:
        return 'Helårsgrund'
    elif int(code) == 401:
        return 'Kolonihave'
    elif int(code) == 600:
        return 'Landejendom'
    elif int(code) == 1400:
        return 'Lystejendom'
    elif int(code) == 100:
        return 'Villa'
    elif int(code) == 900:
        return 'Villalejlighed'
```

```python
df ['bolig_type'] = df['itemType'].apply(xtract_proptype)
```

```python
import string
def remove_signs(string_input):
    list_of_signs = [',', '+', '-',' ','(', '/', ')', '.']
    letters = string.ascii_letters
    found_sign = False
    lowest_value = 100000
    for i in range(len(string_input)):
        if (string_input[i] in list_of_signs) | (string_input[i] in letters):
            found_sign = True
            if i < lowest_value:
                lowest_value = i
                lst = string_input.split(string_input[i])
            return lst[0]
    if found_sign == False:    
        return string_input
#remove_signs('92A, 1. mf')
```

```python
df['number'] = df['number'].astype(str)
```

```python
df['number_plain'] = df['number'].apply(remove_signs)
```
